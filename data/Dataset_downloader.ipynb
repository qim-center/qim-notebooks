# Dataset Downloader
How to download datasets from the Qim data repository.


```python
import qim3d

downloader = qim3d.io.Downloader()
vol = downloader.Cowry_Shell.Cowry_DOWNSAMPLED(load_file=True)
```

```python
qim3d.viz.slices_grid(vol, color_bar=True)
```

```python
qim3d.viz.volumetric(vol, vmin=9000, vmax=50000)
```
