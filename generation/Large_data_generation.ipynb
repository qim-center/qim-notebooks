```python
import qim3d

vol_size = 2000

# Generate synthetic blob
vol = qim3d.generate.volume(final_shape=(vol_size,vol_size,vol_size), noise_scale = 0.02)
```

```python
# Apply noise to the synthetic collection
vol_noisy = qim3d.generate.background(
    background_shape = vol.shape,
    min_noise_value = 0,
    max_noise_value = 20,
    generate_method = 'add',
    apply_method = 'add',
    apply_to = vol
)
```

```python
qim3d.viz.slices_grid(vol_noisy)
```

```python
import qim3d.io
```

```python
qim3d.io.export_ome_zarr(f"vol_{vol_size}.zarr", vol_noisy, chunk_size=100)
```
