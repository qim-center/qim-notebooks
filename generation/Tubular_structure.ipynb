# Tubular structure
In this notebook we generate a tubular structure

```python
import qim3d

vol = qim3d.generate.volume(base_shape = (200, 100, 100),
                        final_shape = (400, 100, 100),
                        noise_scale = 0.03,
                        gamma = 0.12,
                        threshold = 0.85,
                        shape = "tube",
                        )
```

```python
qim3d.viz.slices_grid(vol, num_slices=5, slice_axis=0)
```

```python
qim3d.viz.slices_grid(vol, num_slices=5, slice_axis=1)
```

```python
qim3d.viz.volumetric(vol)
```
